# 1 Data Preparation

**1.1 Merge data from each month into one CSV and read in updated DataFrame**

```python
#Import os module
import os

#Import pandas library
import pandas as pd
```

```python
#define the directory path where all the files are located
directory_path="Sales Analysis 2019/"
```

```python
#List all the files that are in the "Sales Analysis 2019" directory in the form of list
file_list=os.listdir(directory_path)
file_list     #either it will be .csv or .docx files, it display all the files 
```

```python
#Filter out non-CSV files  i.e only .CSV file is added
#Create an empty list named as csv_files
csv_files_2019=[]
for file in file_list: # for loop in the file_list
    if file.endswith('.csv'):  #search for the .csv extension
        csv_files_2019.append(file) #add the .CSV file(s) in csv_files list 
csv_files_2019
```

```python
#Create an empty DataFrame "merged_data_12Months_2019" to store all the 12 months data
merged_data_12Months_2019=pd.DataFrame()
for file in csv_files_2019: #for loop in the csv_files
    file_path=os.path.join(directory_path,file)  #Eg Sales Analysis 2019/Sales_January_2019
    data=pd.read_csv(file_path)  #Retrive the data from CSV files
    #Concatination of heavy lifting files
    merged_data_12Months_2019=pd.concat([merged_data_12Months_2019,data], ignore_index=True)  
merged_data_12Months_2019
```

```python
'''check if "merged_data_12Months_2019.csv" exists or not
if exists run if statement. if not then run else statement'''

if os.path.exists('merged_data_12Months_2019.csv'):   
    print("merged_data_12Months_2019.csv already exists !!!")
else: 
    
    #Export and save merged_data_12Months_2019 DataFrame to a CSV file named "merged_data_12Months_2019.csv" 
    merged_data_12Months_2019.to_csv('merged_data_12Months_2019.csv', index=False) #ignore index
    print("Merged data of 12 months saved to merged_data_12Months_2019.csv")

```

**1.2 Remove the NaN missing values from the updated dataframe**

***Checking missing vallues***

```python
#isna() function to detect missing values
NaNdf=merged_data_12Months_2019.isna()

#view 258 rows sample of Dataframe 
NaNdf.head(258) 
```

***Droping missing values***

```python
#remove row(s) from a Dataframe
merged_data_12Months_2019=merged_data_12Months_2019.dropna()

#View 258 rows sample of Dataframe
merged_data_12Months_2019.head(258)
```

**1.3 Convert "Quantity Ordered" and "Price Each" to numeric**

***Show the datatypes of columns***

```python
print(merged_data_12Months_2019.dtypes)
```

***Converting "Quantity Ordered" and "Price Each" columns to numeric***

```python
#Using dictionary to convert the data type of  specific colums
convert_dict={'Quantity Ordered': int,
              'Price Each': float   
             }

#assign convert_dict as parameter. 
merged_data_12Months_2019=merged_data_12Months_2019.astype(convert_dict)
print(merged_data_12Months_2019.dtypes)
```

**1.4. Create a new column named Month from Ordered Date of updated dataframe and convert it to integer as data type.**

```python
#Display all the columns
merged_data_12Months_2019.columns 
```

```python
#Convert "Order Date" to datetime as datatype
merged_data_12Months_2019['Order Date']=pd.to_datetime(merged_data_12Months_2019['Order Date'])

#Extract month from "Ordered Date"
merged_data_12Months_2019['Month']=merged_data_12Months_2019['Order Date'].dt.month
#convert "Month" to integer
merged_data_12Months_2019['Month']=merged_data_12Months_2019['Month'].astype(int)

print(merged_data_12Months_2019.dtypes)
merged_data_12Months_2019



```

**1.5 Create a new column named City from Purchase Address based on the value in updated dataframe.**

```python
#Create an empty list to store extracted city names from "Purchase Address"
city_list=[]

#Loop through each address in the "Purchase Address" column
for address in merged_data_12Months_2019['Purchase Address']: 
    
    #Split the address by comma and extract the element at the index 1 i.e city name
    city=address.split(',')[1]
    
    #add the extracted city name in city_list
    city_list.append(city)
#Assign the city_list to a new "City" column in the dataframe
merged_data_12Months_2019['City']=city_list
merged_data_12Months_2019
```

# 2 Data Analysis

**2.1 Write a Python program to show summary statistics of sum, mean, standard deviation, skewness, and kurtosis of any chosen variable**

```python
#Assigning 'Quantity Ordered' column values in a quantityOrdered variable
quantityOrdered=merged_data_12Months_2019['Quantity Ordered']
#type(quantityOrder)   #Series

#sum of all data in quantiyOrdered Series
print("The sum of quantity ordered is: ",quantityOrdered.sum())

#calculates the mean(average) of quantityOrdered Series
print("The mean of quantity ordered is: ",quantityOrdered.mean())

#Calculates the standard deviation of quantityOrdered Series
print("The Standard Deviation of quantity ordered is: ",quantityOrdered.std())

print("The skewness of quantity ordered is: ",quantityOrdered.skew())

print("The kurtosis of quantity ordered is: ",quantityOrdered.kurtosis())
```

# 3 Data Exploration

**3.1 Which Month has the best sales? and how much was the earning in that month? Make a bar graph of sales as well.**

```python
#Add 'Total Sales' column
#Calculate total sales of each day
# Since, the right side is not visible so the rest of code is... * merged_data_12Months_2019['Price Each'] 
merged_data_12Months_2019['Total Sales']=merged_data_12Months_2019['Quantity Ordered']*merged_data_12Months_2019['Price Each']
merged_data_12Months_2019.head()
```

```python
#group by months and add the sales of each month
monthly_sales=merged_data_12Months_2019.groupby('Month', as_index=False).sum()
monthly_sales
```

```python
#find the max value in 'Total Sales' Column and get the entire column
max_sales=monthly_sales.loc[monthly_sales['Total Sales'].idxmax()] 
max_sales

```

```python
import matplotlib.pyplot as plt

```

***Visualizing sales in different months***

```python
months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sept','Oct','Nov','Dec'] #for x axis
results=merged_data_12Months_2019.groupby('Month').sum() 
sales=results['Total Sales']

#create a bar graph
plt.bar(months,sales, color="black")

#Add labels and title
plt.xlabel("Months")
plt.ylabel("sales in USD")
plt.title("Sales in different Months in 2019/Sales Distribution", color="maroon")

#Change the SclarFormatter of y axis
plt.ticklabel_format(axis='y',style='plain')  # removing le6 from bar graph

#show the graph
plt.show()

```

**3.2 Which city has sold the highest product?**

```python
#group by city and add the values based on group
city_sales=merged_data_12Months_2019.groupby('City',as_index=False)[['Quantity Ordered','Total Sales']].sum()
city_sales
```

```python
#Filter the row with the maximum sales
highest_sales=city_sales[city_sales['Total Sales']==city_sales['Total Sales'].max()]
highest_sales   
```

**3.3 Which product was sold the most in overall? Illustrate it through bar graph.**

```python
#group by product and add the each values
product_sales=merged_data_12Months_2019.groupby('Product',as_index=False)[['Quantity Ordered','Total Sales']].sum()
product_sales
```

```python
#Filter the row with the maximum sales
highest_sales_product=product_sales[product_sales['Quantity Ordered']==product_sales['Quantity Ordered'].max()]
highest_sales_product
```

***Visualizing the result in bar graph***

```python
products=product_sales['Product']
#quantity=product_sales['Quanity Ordered']
quantity_ordered=product_sales['Quantity Ordered']

plt.bar(products,quantity_ordered, color="black")

plt.xlabel("Products")
plt.ylabel("Quantity Ordered")
plt.title("Quantity Ordered of different products",color="maroon")

plt.ticklabel_format(axis='y',style='plain') 
plt.xticks(products,rotation='vertical', size=8)

plt.show()
```
